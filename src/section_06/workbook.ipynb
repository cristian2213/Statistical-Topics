```python
from stats import Statistics

# ******************************
# DISCRETE PROBABILITY
# ******************************
# 2
data = [(0, 1 / 5), (5, 1 / 5), (10, 2 / 5), (15, 1 / 5)]
res = Statistics.get_discrete_variance(data)
expected_value = res[1]
variance = res[0]
std = Statistics.get_discrete_std(variance)
print(
    "#2",
    "\n",
    "mean ->",
    expected_value,
    "\n",
    "variance ->",
    variance,
    "\n",
    "std ->",
    std,
)

# 3
household_3 = 1 - (0.281 + 0.340 + 0.129 + 0.060 + 0.023 + 0.013)
p_at_least_4 = 1 - (0.281 + 0.340 + household_3)  # P(X>=4) = 1 - P(X<4)
p_at_least_2 = 1 - 0.28  # P(X>=2) = 1 - P(X=1)
p_at_least_4_given_at_least_2 = p_at_least_4 / p_at_least_2
print(
    "#3",
    "\n",
    "p_at_least_2 ->",
    p_at_least_2,
    "\n",
    "p_at_least_4 ->",
    p_at_least_4,
    "\n",
    "p_at_least_4_given_at_least_2 ->",
    p_at_least_4_given_at_least_2,
)

# 4
# total cards = 52
# red cards = 26
# black cards = 26
# R = red
# B = black
deck = 52
p_0_red = (26 / 52) * (25 / 51)  # P(BB)
p_2_red = (26 / 52) * (25 / 51)  # P(RR)
p_1_red = 1 - p_2_red - p_0_red  # P(RB) = P(RB) + P(BR)
p_1_red_alternative = ((26 / 52) * (26 / 51)) + ((26 / 52) * (26 / 51))
print(
    "#4",
    "\n",
    "p_0_red ->",
    p_0_red,
    "\n",
    "p_1_red ->",
    p_1_red,
    "\n",
    "p_1_red_alternative ->",
    p_1_red_alternative,
    "\n",
    "p_2_red ->",
    p_2_red,
)

# 5
"""
A local restaurant features a wheel we can spin before paying the bill.
The wheel is split into 8 equal size pieces. One of the sections gives us a
$10 discount on the bill, two sections give a $5 discount, three sections give
a $2 discount, and the rest of the sections give no discount. Find the
expected value for the discount given by the wheel.
"""
slice_1 = (10, 1 / 8)
slice_2 = (5, 2 / 8)
slice_3 = (2, 3 / 8)
slice_4 = (0, 2 / 8)
data = [slice_1, slice_2, slice_3, slice_4]
expected_value = Statistics.get_discrete_mean(data)
print("#5", "\n", "mean ->", expected_value)

# 6
"""
John stops at the local gas station and decides to buy lottery tickets.
Each ticket has a 20 % chance of being a winner. He will buy a lottery ticket
and check to see if it’s a winner. If it’s a winner, he’ll collect his money and
be done. If it’s not a winner, he’ll buy another. He’ll repeat this until he gets
a winning ticket. But if he hasn’t won by his fifth ticket, he won’t buy any
more tickets. Let L be the number of lottery tickets John will buy, then find E(L).
"""
p_win = 0.2
trials = 5
p_l_1 = Statistics.get_geometric_probability(1, p_win)
p_l_2 = Statistics.get_geometric_probability(2, p_win)
p_l_3 = Statistics.get_geometric_probability(3, p_win)
p_l_4 = Statistics.get_geometric_probability(4, p_win)
p_l_5 = 1 - (p_l_1 + p_l_2 + p_l_3 + p_l_4)  # P(L=5) = 1 - P(L<5) = 1 - P(L<=4)
expected_value = sum([i * j for i, j in enumerate([p_l_1, p_l_2, p_l_3, p_l_4, p_l_5], 1)])
print(
    "#6",
    "\n",   
    "p_l_1 ->",
    p_l_1,
    "\n",
    "p_l_2 ->",
    p_l_2,
    "\n",
    "p_l_3 ->",
    p_l_3,
    "\n",
    "p_l_4 ->",
    p_l_4,
    "\n",
    "p_l_5 ->",
    p_l_5,
    "\n",
    "expected_value ->",
    expected_value,
)
```
